**1. Charger un graphe .net (Pajek) avec NetworkX.**

```python
import kagglehub
import os, glob
import networkx as nx

# Télécharger dataset de graphes
path = kagglehub.dataset_download("mateuscco/toy-network-datasets")
print("Dataset téléchargé dans :", path)

# Chercher un fichier Pajek (.net)
graph_files = glob.glob(os.path.join(path, "**", "*.net"), recursive=True)
print("Fichiers trouvés :", graph_files)

# Charger un graphe Pajek
G = nx.read_pajek(graph_files[0])
print("Graphe chargé :", graph_files[0])
```

**2. Afficher le nombre de nœuds et d’arêtes.**

```python
print("Nombre de nœuds :", G.number_of_nodes())
print("Nombre d’arêtes :", G.number_of_edges())
```

**3. Visualiser le graphe avec matplotlib.**

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
nx.draw(G, with_labels=True, node_color="skyblue", edge_color="gray", node_size=500, font_size=10)
plt.title("Visualisation du graphe")
plt.show()
```

**4. Calculer le degré de chaque nœud.**

```python
degrees = dict(G.degree())
print("Degré des nœuds :", degrees)
```

***5. Tracer la distribution des degrés (histogramme).***

```python
plt.figure(figsize=(6,4))
plt.hist(list(degrees.values()), bins=range(1, max(degrees.values())+2), align="left", color="orange", edgecolor="black")
plt.title("Distribution des degrés")
plt.xlabel("Degré")
plt.ylabel("Nombre de nœuds")
plt.show()
```
